```python
import numpy as np
from datascience import *
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('fast')
import random
from numpy import random
import math
import random
```

```python
def molecules_adsorbed (T,ep):
    hJ = 6.626e-34
    Planck_constant = (hJ)**3
    massH2 = 3.32e-27
    mass2pi = (massH2*np.pi*2)**(3/2)
    Kb = 1.38e-23
    KbT = Kb*T
    temp = (KbT)**(5/2)
    beta = 1/(KbT)
    PqT = (mass2pi*temp)/Planck_constant
    epsilonJ = (ep*1000)/6.022e23
    exponential = PqT*np.exp(-1*beta*epsilonJ)
    return 101325/(101325+exponential)
```

```python
temp300 = []
for i in np.arange(1,500,0.25):
    temp300.append(molecules_adsorbed (300,i))
theta300 = np.array(temp300)
```

```python
temp300
```

```python
temp400 = []
for i in np.arange(1,500,0.25):
    temp400.append(molecules_adsorbed (400,i))
theta400 = np.array(temp400)
```

```python
temp400
```

```python
epsilon = np.linspace(1,500,1996)
```

```python
difference = np.subtract(temp300,temp400)
difference
```

```python
table = Table().with_columns("epsilon",epsilon,"300",temp300,"400",temp400,"difference",difference)
table.plot(0)
```

```python
# this is the range where the change is greater than 0.5
# 29.5143KJ/mol - 40.7699 KJ/mol
range_difference = table.select("epsilon","difference").where("difference", are.above(0.5))
range_difference.show(0)
```
